```python
# Print the actual column names to identify any discrepancies
print(heart_dataset.columns)  

# Assuming the column name is 'target' based on common naming conventions 
# in heart disease datasets, adjust as needed
x = heart_dataset.drop(columns='target', axis=1)  
y = heart_dataset['target']
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score # Import necessary metrics

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train) # Changed 'X_train' to 'x_train'
y_pred_rf = rf_model.predict(x_test) # Changed 'X_test' to 'x_test'

# Evaluate
print("\nRandom Forest:")
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("R² Score:", r2_score(y_test, y_pred_rf))
```

Using Decision Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize Decision Tree Classifier
model = DecisionTreeClassifier(random_state=42)

# Train the model
# Changed 'X_train' to 'x_train' to match the variable defined earlier
model.fit(x_train, y_train)
```

```python
# Predict on test data
y_pred = model.predict(x_test)  # Changed 'X_test' to 'x_test'
```

```python
# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```
